```python
# all trasaction in list
# non null values for all tha colum
# orvaru colum la erkaru non null  value edtu list la append pani erkanum

transactions = []
for i in range(0, len(d)):transactions.append([str(d.values[i,j]) for j in range(0,20)if str(d.values[i,j])!='0'])
```

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

```python
result=list(rules)
```

```python
result
```

```python
# fetch panannum dataframe ha add paan
lhs= []
rhs = []
supports = []
confidences = []
lifts = []

for result in result:
  for statistic in result.ordered_statistics:
    lhs.append(list(statistic.items_base))
    rhs.append(list(statistic.items_add))
    supports.append(result.support)
    confidences.append(statistic.confidence)
    lifts.append(statistic.lift)

res = pd.DataFrame({
    "lhs":lhs,
    "rhs":rhs,
    "support":supports,
    "confidence":confidences,
    "lift":lifts
})
res
```
