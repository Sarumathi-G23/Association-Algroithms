```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df = pd.read_csv("/content/OnlineRetail.csv", encoding='ISO-8859-1')
```

```python
print("Dataset shape:", df.shape)
print(df.head())
```

```python
df = df[df['InvoiceNo'].notnull() & df['Description'].notnull()]
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df['Description'] = df['Description'].str.strip()
```

```python
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
         .set_index('InvoiceNo'))
```

```python
# Convert quantities to 1/0 (presence/absence)
def encode_units(x):
    return 1 if x > 0 else 0

basket_sets = basket.applymap(encode_units)
```

```python
# 5. Apply FP-Growth
frequent_itemsets = fpgrowth(basket_sets, min_support=0.02, use_colnames=True)
```

```python

print("\nFrequent Itemsets:")
print(frequent_itemsets.sort_values(by='support', ascending=False).head(10))
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
```
